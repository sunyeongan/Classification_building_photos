```python
# 주어진 디렉토리에 있는 항목들의 이름을 담고 있는 리스트를 반환합니다.
# 리스트는 임의의 순서대로 나열됩니다.
# 폴더 순환 해서 이미지이름이 바뀌지 않은 것들은 모두 일괄로 변경한다. 
path = '/content/drive/MyDrive/건물사진분류/'


dirnames = ['0','1','2','3','4']
dirnames2 = [0,1,2,3,4]
dir=[]


for i in range(0,5):
  dir_path = path + str(i)
  dir.append(dir_path)
  #print(dir)
  
  for j in dir:
    file_names = os.listdir(j)
    #print(file_names)
    cnt=0

    for k in file_names:
      if k[0] =='C':#폴더안 이름이 이미 바뀐 이미지의 갯수를 가져와 cnt에 저장
        cnt += 1

      elif k[0] != 'C':
        print(k)
        src = os.path.join(j,k)
        #print(src)
        dst = 'C0'+str(i)+'_SY_'+ datetime.datetime.now().strftime('%Y%m%d_')+ str(cnt).zfill(3) + '.jpg'
        
        dst = os.path.join(j, dst)
        #print(dst)
        os.rename(src, dst)
        cnt += 1

```
