```python
import os
import datetime
from PIL import Image
import pyheif
```

```python
path = '/content/drive/MyDrive/건물사진분류/'


dirnames = ['0','1','2','3','4']
dirnames2 = [0,1,2,3,4]



```

```python
dir=[]
for i in range(5):
  dir_path = path + str(i)
  dir.append(dir_path)
  print(dir)


for j, file_path in enumerate(dir):
  file_names = os.listdir(file_path)
  print(j)
  cnt=0
  for k in file_names:
    if k[0] =='C':#폴더안 이름이 이미 바뀐 이미지의 갯수를 가져와 cnt에 저장
      cnt += 1

    elif k[0] != 'C':
      print(k)
      src = os.path.join(file_path,k)
      #print(src)
      heif_file = pyheif.read(src)
      image = Image.frombytes(
            heif_file.mode, 
            heif_file.size, 
            heif_file.data,
            "raw",
            heif_file.mode,
            heif_file.stride,
            )
      #print(heif_file)
      dst = 'C0'+str(j)+'_SY_'+ datetime.datetime.now().strftime('%Y%m%d_')+ str(cnt).zfill(3)
      
      dst = os.path.join(file_path, dst)
      image.save(dst+".jpg","JPEG")    
      print(dst)
      os.rename(src,dst)
      cnt += 1

```

```python
#.jpg 확장자가 아닌것들은 모두 지운다.

for i in range(0,5):
  dir_path = path + str(i)
  dir.append(dir_path)
  #print(dir)
  
  for j in dir:
    file_names = os.listdir(j)
    print(file_names)
    cnt=0

    for k in file_names:
      if k[0] =='C':#폴더안 이름이 이미 바뀐 이미지의 갯수를 가져와 cnt에 저장
        cnt += 1
        if k[-4:] != ".jpg":
          file_path = os.path.join(j,k)
          #print(file_path)
          os.remove(file_path)

```
